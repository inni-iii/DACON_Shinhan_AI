```python
# selenium 4
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd

# selenium으로 키를 조작하기 위한 import
from selenium.webdriver.common.keys import Keys

# 페이지 로딩을 기다리는데에 사용할 time 모듈 import
import time
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# 신한카드 신한플레이 이벤트 페이지(진행중)
driver.get('https://www.shinhancard.com/mob/MOBFM026N/MOBFM026C01.shc')

# 자동 저장을 위한 타이틀 폴더
title_folder = './image'

if not os.path.isdir(title_folder):
    os.mkdir(title_folder)

# 신한플레이 이벤트
titles = driver.find_elements(By.CSS_SELECTOR, 'div.img-sec')
shinhanplay_event = []
for idx, title in enumerate(titles):
    text = title.text
    shinhanplay_event.append(text)

df = pd.DataFrame(shinhanplay_event, columns=['title'])
```

```python
# 첫 번째 컬럼은 다른 정보라서 지우기
df.drop([0], axis=0, inplace=True)
```

```python
df
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# 신한카드 신한플레이 이벤트 페이지(종료)
driver.get('https://www.shinhancard.com/mob/MOBFM026N/MOBFM026C10.shc')

shinhanplay_event = []
while True:
    # 신한플레이 이벤트(종료)
    titles = driver.find_elements(By.CSS_SELECTOR, 'a.title')

    for idx, title in enumerate(titles):
        text = title.text
        shinhanplay_event.append(text)
  
    try:
        driver.find_element(By.CSS_SELECTOR, 'button#prevBtn.swiper-button-prev').click()
    except:
        break
    
    time.sleep(3)

df2 = pd.DataFrame(shinhanplay_event, columns=['title'])
```

```python
df2
```

```python
df3 = pd.concat([df, df2])
df3
```

```python
df3.to_csv('./title/shinhanplay_event.csv', index=False)
```
